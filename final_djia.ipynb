```python
import datetime
import numpy as np
from matplotlib import cm, pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn import tree
from pandas_datareader import data
import ta
```

```python
t1=pd.read_csv("dtf.csv").iloc[7050:-1]
t2=pd.read_csv("tf.csv").iloc[7050:-1]
t3=pd.read_csv("cnn.csv").iloc[7050:-1]
```

```python
t41=((t1.Close-24777)/24777)*100
```

```python
from matplotlib import style
style.use('ggplot')
```

```python
%matplotlib inline
# Control the default size of figures in this Jupyter notebook
%pylab inline
fig = plt.figure()
pylab.rcParams['figure.figsize'] = (15, 15)  
plt.plot(t11,'r',label='Proposed Model',linewidth=1)
plt.plot(t21,'g',label='Trend Following', linewidth=1)
plt.plot(t31,'b',label='CNN',linewidth=1)
plt.plot(t41,'c',label='Buy & Hold',linewidth=1)
#plt.title('Epic Info')
plt.ylabel('% Accumulated Return')
plt.xlabel('Testing Period (days)')
plt.legend()
plt.grid(True,color='k')
#plt.gca().axes.get_xaxis().set_visible(False)
plt.show()
fig.savefig('DJIA_final.png')
```
